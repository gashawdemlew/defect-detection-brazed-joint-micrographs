```python
import os
import cv2
import glob
from ultralytics import YOLO
import matplotlib.pyplot as plt

# --- Configuration ---

# --- Configuration ---

# 1. LOAD YOUR TRAINED MODEL
# IMPORTANT: Replace this path with the actual path to your 'best.pt' file
# It's typically found in the 'runs/segment/train/weights/' directory.
MODEL_PATH = '/content/defect_detector_model.pt'

# 2. DEFINE PATHS
# Folder containing the images you want to run inference on.
IMAGE_FOLDER_PATH = '/content/sgementation-brazed-joint-2/train/images'

# Folder where the results (images with segmentation) will be saved.
OUTPUT_FOLDER_PATH = 'results/'

# --- End of Configuration ---


# Create the output directory if it doesn't exist
os.makedirs(OUTPUT_FOLDER_PATH, exist_ok=True)

# Check if the model file exists
if not os.path.exists(MODEL_PATH):
    print(f"Error: Model file not found at '{MODEL_PATH}'")
    print("Please make sure you have provided the correct path to your 'best.pt' file.")
else:
    # Load the trained YOLOv11 model
    print(f"Loading model from {MODEL_PATH}...")
    model = YOLO(MODEL_PATH)
    print("Model loaded successfully.")

    # Get a list of all image files in the folder (jpg, png, jpeg)
    image_files = []
    for ext in ['*.jpg', '*.jpeg', '*.png']:
        image_files.extend(glob.glob(os.path.join(IMAGE_FOLDER_PATH, ext)))

    # --- MODIFICATION: LIMIT TO THE FIRST 10 IMAGES ---
    # After finding all images, take only the first 10 from the list.
    if len(image_files) > 10:
        print(f"Found {len(image_files)} total images. Processing the first 10.")
        image_files = image_files[10:20]
    # --- END OF MODIFICATION ---

    if not image_files:
        print(f"No images found in '{IMAGE_FOLDER_PATH}'. Please check the path and file extensions.")
    else:
        print(f"Processing {len(image_files)} images.")

        # Loop through the limited list of image files
        for image_path in image_files:
            print(f"\nProcessing: {os.path.basename(image_path)}")

            # Run inference on the image
            results = model(image_path)

            # The .plot() method returns a BGR numpy array with detections and masks plotted
            result_image_array = results[0].plot()

            # --- Display the result ---
            # Convert the BGR image (from OpenCV) to RGB for Matplotlib
            result_image_rgb = cv2.cvtColor(result_image_array, cv2.COLOR_BGR2RGB)

            plt.figure(figsize=(10, 10))
            plt.imshow(result_image_rgb)
            plt.title(f'Segmentation Result for {os.path.basename(image_path)}')
            plt.axis('off')
            plt.show()

            # --- Save the result ---
            output_filename = f"result_{os.path.basename(image_path)}"
            output_path = os.path.join(OUTPUT_FOLDER_PATH, output_filename)
            cv2.imwrite(output_path, result_image_array)
            print(f"Result saved to: {output_path}")

print("\nInference complete.")
```
